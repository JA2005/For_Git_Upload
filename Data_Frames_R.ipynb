# What is data frame in R?

# Data Frame

* It is a fundamental date structur to start a dataset
* It is similar to matrix, as it has rows and columns 
* Rows corresponds to observations and columns corresponds to variables
* Contains elements of different types, one column may contain numerical , and another one may contain character variable
* Data frame is generated by function data.frame()
* Lets start our practical work now!!

```r
my_df <- data.frame(
names = c("Susan", "Jimmy", "Jack", "Tony", "Matt"),
marks = c(80, 30, 78, 90, 88),
result = c("Pass", "Fail", "Pass", "Pass", "Pass")
)


my_df
```

```r
rownames(my_df) <- c("row1", "row2", "row3","row4", "row5") # for naming rows 
## similarly column names can also be given by rownames(my_df) <- c("col1", "col2", "col3","col4", "col5")
my_df
```

```r
my_df[1,2] # row 1 column 2
```

```r
my_df[2,1] # row 2 col 1
```

```r
my_df[,1] # all row col 1
```

```r
my_df[1,] #  row 1 col all
```

```r
# $ can also be used for that 
my_df$names
```

```r
my_df$marks # another way
```
